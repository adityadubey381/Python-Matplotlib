```python
df = pd.read_excel(r"C:\Users\Aditya kumar Dubey\Downloads\IMdB final file.xlsx\IMdB final file.xlsx")
df.head(2)
```

```python
df.columns
```

```python
dfs = df[['Movie Key', 'Release_Year', 'Title', 'IMdB Rating','Number of Votes','Certification_Rating', 'RT (mins)', 'Release_Day', 'Release_Month',
       'Boxoffice_Collection ($ USD)', 'User_Reviews']]
```

```python
dfs.shape
```

```python
dfs.columns
```

```python
dfs.shape
```

```python
top10 = dfs.sort_values(by = 'IMdB Rating',ascending = False).head(10)

plt.bar(top10['Title'], top10['Number of Votes'],color = 'black')
plt.xticks(rotation = 'vertical')
plt.show()
```

```python
#How has the number of movies released per year changed over time?
year = dfs.groupby('Release_Year').size().reset_index(name='Count')
plt.plot(year['Release_Year'],year['Count'], marker='o', linestyle='-', color='skyblue')
plt.xlabel('years')
plt.ylabel('Number of Movie')
plt.tilte("Movies Released Per Year")
plt.grid()
plt.show()
```

```python

plt.figure(figsize=(12, 6))
plt.plot(year['Release_Year'], year['Count'], marker='o', linestyle='-', color='skyblue')
plt.title("Movies Released Per Year")
plt.grid()
plt.show()
```
