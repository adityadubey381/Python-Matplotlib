```python
plt.plot(batsman['index'],batsman["V Kohli"])
plt.plot(batsman['index'],batsman["RG Sharma"])
```

```python
#lable title
plt.plot(batsman['index'],batsman["V Kohli"])
plt.plot(batsman['index'],batsman["RG Sharma"])

plt.title("ham rohit sharma or virst kohli ka career Comprision kar sahe hai")
plt.xlabel("season")
plt.ylabel("Runs Scored")
```

```python
#colors(hex)  and line (width and style) and marker(size)
plt.plot(batsman['index'],batsman["V Kohli"],color = "orange")
plt.plot(batsman['index'],batsman["RG Sharma"],color = "green")

plt.title("ham rohit sharma or virst kohli ka career Comprision kar sahe hai")
plt.xlabel("season")
plt.ylabel("Runs Scored")
```

```python
plt.plot(batsman['index'],batsman["V Kohli"],color = "orange", linestyle = "dotted")
plt.plot(batsman['index'],batsman["RG Sharma"],color = "green",linestyle ="dotted" )

plt.title("ham rohit sharma or virst kohli ka career Comprision kar sahe hai")
plt.xlabel("season")
plt.ylabel("Runs Scored")
```

```python
plt.plot(batsman['index'],batsman["V Kohli"],color = "orange", linestyle = "dashdot" ,linewidth = 4)
plt.plot(batsman['index'],batsman["RG Sharma"],color = "green",linestyle ="solid", linewidth = 3)

plt.title("ham rohit sharma or virst kohli ka career Comprision kar sahe hai")
plt.xlabel("season")
plt.ylabel("Runs Scored")
```

```python
#Marker
plt.plot(batsman['index'],batsman["V Kohli"],color = "#33FF88", linestyle = "dashdot" ,linewidth = 4, marker = '*')
plt.plot(batsman['index'],batsman["RG Sharma"],color = "#FF9B33",linestyle ="solid", linewidth = 3,marker = "D",markersize = 5)

plt.title("ham rohit sharma or virst kohli ka career Comprision kar sahe hai")
plt.xlabel("season")
plt.ylabel("Runs Scored")
```

```python
plt.plot(batsman['index'],batsman["V Kohli"],color = "#33FF88", linestyle = "dashdot" ,linewidth = 4, marker = 'D', markersize = 10)
plt.plot(batsman['index'],batsman["RG Sharma"],color = "#FF9B33",linestyle ="solid", linewidth = 3,marker = "o",markersize = 5)

plt.title("ham rohit sharma or virst kohli ka career Comprision kar sahe hai")
plt.xlabel("season")
plt.ylabel("Runs Scored")
```

```python
#legend -> location
plt.plot(batsman['index'],batsman["V Kohli"],color = "#33FF88", linestyle = "dashdot" ,linewidth = 4, marker = 'D', markersize = 10,label = "virat")
plt.plot(batsman['index'],batsman["RG Sharma"],color = "#FF9B33",linestyle ="solid", linewidth = 3,marker = "o",markersize = 5,label = "Rohit")

plt.title("ham rohit sharma or virst kohli ka career Comprision kar sahe hai")
plt.xlabel("season")
plt.ylabel("Runs Scored")
plt.legend()
```

```python
#limiting axes
price = [48000,54000,57000,49000,47000,45000,45000000]
year = [2015,2016,2017,2018,2019,2020,2021]

plt.plot(year,price)
plt.ylim(0,100000)
```
